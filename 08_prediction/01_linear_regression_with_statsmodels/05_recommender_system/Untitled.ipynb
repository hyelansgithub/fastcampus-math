```python
import surprise
import pandas as pd
```

```python
data = surprise.Dataset.load_builtin('ml-100k')
```

```python
df = pd.DataFrame(data.raw_ratings, columns=["user", "item", "rate", "id"])
del df["id"]
df.head(10)
```

```python
df_table = df.set_index(["user", "item"]).unstack()
```

## 4. 누적 분포 함수와 확률 밀도 함수

- 표본 : 하나의 경우
- 표본 집단 : 가능한 경우의 전체
- 사건 : 표본 집단의 부분집합
- 확률 : 사건에 대해 숫자를 할당하는 함수
- 확률 변수 : 표본에 실수값을 부여하는 함수

- - -

- 확률 분포 : 각 사건에 할당된 확률의 분포를 나타내는 것
    - 모든 사건에 대한 확률을 할당하는 것이 쉽지 않아서 구간을 이용해서 묘사한다.
    - 시작점을 -∞로 동일하게 작성하면 간단히 작성 가능
- - -
- 누적 분포 함수 (CDF) : -∞에서 확률 변수까지의 누적된 확률을 보여준다.
- - -
- 확률 밀도 함수 (PDF) : 상대적인 확률의 변화를 직관적으로 볼 수 있게 도와줌
- 확률 질량 함수 (PMF): 이산 확률 변수의 값 하나 하나에 대한 확률을 정의

- - -
### 1. 누적 분포 함수

```python
import numpy as np
import matplotlib.pylab as plt

t = np.linspace(-100, 500, 100)
F = t / 360
F[t < 0] = 0
F[t > 360] = 1
plt.plot(t, F)
plt.ylim(-0.1, 1.1)
plt.xticks([0, 180, 360])
plt.title("Cumulative Distribution Function")
plt.xlabel("$x$ (deg.)")
plt.ylabel("$F(x)$")
plt.show()
```

- - -
### 2-1. 확률 밀도 함수

```python
t = np.linspace(-100, 500, 1000)
F = t / 360
F[t < 0] = 0
F[t > 360] = 1
f = np.gradient(F, 600/1000)  # 수치미분
plt.plot(t, f)
plt.ylim(-0.0001, f.max()*1.1)
plt.xticks([0, 180, 360])
plt.title("Probability Density Function")
plt.xlabel("$x$ (deg.)")
plt.ylabel("$f(x)$")
plt.show()
```

### 2-2. 확률 질량 함수

```python
x = np.arange(1,7)
y = np.array([0.1, 0.1, 0.4, 0.2, 0.1, 0.1])
plt.stem(x, y)
plt.xlim(0, 7)
plt.ylim(-0.01, 0.5)
plt.show()
```

```python
x = np.arange(1,7)
y = np.array([0.1, 0.1, 0.4, 0.2, 0.1, 0.1])
z = np.cumsum(y)
plt.step(x, z, where="post");
plt.xlim(0, 7);
plt.ylim(-0.01, 1.1)
plt.show()
```

## 보스턴 주택가격 데이터
- - -
bunch안에 data, target, feature_name, target_name, DESCR 5개 중 있는 것도 있고 없는 것도 있음!

    - target_name : 1978 보스턴 주택 가격, 506 타운의 주택 가격 중앙값 (단위 1,000 달러)
    
    - feature_name 
        - CRIM: 범죄율
        - INDUS: 비소매상업지역 면적 비율
        - NOX: 일산화질소 농도
        - RM: 주택당 방 수
        - LSTAT: 인구 중 하위 계층 비율
        - B: 인구 중 흑인 비율
        - PTRATIO: 학생/교사 비율
        - ZN: 25,000 평방피트를 초과 거주지역 비율
        - CHAS: 찰스강의 경계에 위치한 경우는 1, 아니면 0
        - AGE: 1940년 이전에 건축된 주택의 비율
        - RAD: 방사형 고속도로까지의 거리
        - DIS: 직업센터의 거리
        - TAX: 재산세율

```python
# 데이터 import
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.data
```

```python
boston.target
```

```python
boston.feature_names
```

    target name 없으면 x1,x2,...xn 으로 채워넣기

```python
#존재하지 않음
boston.target_names
```

```python
print(boston.DESCR)
```

```python
# boston bunch 안의 데이터들을 DataFrame으로 만들어주자!
import pandas as pd
dfX = pd.DataFrame(boston.data, columns=boston.feature_names)
dfX.tail()
```

```python
dfy = pd.DataFrame(boston.target, columns=['MEDV'])
dfy.tail()
```

```python
# 보통 독립변수와 종속변수 합쳐서 한 Dataframe으로 만든 후 사용한다!
# unique한 값으로 합치려면 merge사용하고, 단순 옆에 붙이려면 concat 사용

df = pd.concat([dfX,dfy],axis=1)
df.tail()
```

```python
# 데이터의 기술 통계량 분석
df.describe()
```

```python
# 데이터 시각화
import seaborn as sns
import matplotlib.pylab as plt

cols = ["LSTAT","NOX","RM","MEDV"]
sns.pairplot(df[cols])
plt.show()
```

## 당뇨병 데이터
- - -
442 명의 당뇨병 환자의 검사 데이터이다. load_diabetes 명령으로 로드하며 다음과 같이 구성되어 있다.

- 타겟 데이터
    - 1년 후의 당뇨병 진행도
- 특징 데이터
    - 나이, 성병, BMI(body mass index) 지수, 혈압 및 6개의 혈청 검사 수치

```python
from sklearn.datasets import load_diabetes
```

```python
# diabetes bunch만들기
diabetes = load_diabetes()
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
diabetes.feature_names
```

```python
diabetes.target_names
```

```python
print(diabetes.DESCR)
```

```python
dfX = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
dfX.tail()
```

```python
dfy = pd.DataFrame(diabetes.target, columns=["y"])
dfy.tail()
```

```python
df = pd.concat([dfX,dfy],axis=1)
df.tail()
```

```python
df.describe()
```

```python
# pairplot에 직접 컬럼이름 적으면 [[]] 대괄호 두번 적혀야 한다.
sns.pairplot(df[["age","sex","bmi","y"]])
plt.show()
```

## 체력 검사 데이터
- - -
신체 상황 및 운동 능력을 측정한 데이터이다. load_linnerud() 명령으로 로드하며 다음과 같이 구성되어 있다.

    - 턱걸이, 앉았다 일어나기, 점프의 세가지 운동 능력
    - 체중, 허리둘레, 맥박의 세가지 신체 상황

```python
from sklearn.datasets import load_linnerud
```

```python
linnerud = load_linnerud()
```

```python
linnerud.data
```

```python
linnerud.target
```

```python
linnerud.feature_names
```

```python
linnerud.target_names
```

```python
print(linnerud.DESCR)
```

```python
# dataframe으로 만들기
dfX = pd.DataFrame(linnerud.data, columns = linnerud.feature_names)
dfX.tail()
```

```python
dfy = pd.DataFrame(linnerud.target, columns=linnerud.target_names)
dfy.tail()
```

```python
# 두개 df 합치기
df = pd.concat([dfX,dfy],axis=1)
df.tail()
```

```python
df.describe()
```

```python
sns.pairplot(df.iloc[:,[0,1,2,-1]])
plt.show()
```

    - df 개념적으로 dictionary of column series
    - loc
        - loc[숫자 or 문자,숫자or 문자]
        - iloc[숫자, 숫자] : 숫자로 반드시 사용해야
        - ix[] : 문자나 숫자로 많이 사용해도 된다. ix 앞을 없애기로 해서 위에 두개를 사용하는 것이 좋다.

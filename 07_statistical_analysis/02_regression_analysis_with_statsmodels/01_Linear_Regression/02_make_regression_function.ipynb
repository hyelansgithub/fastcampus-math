## 2. 회귀분석용 가상 데이터 생성 방법

- - -


    - Scikit-learn의 datasets 서브 패키지는 회귀분석용 가상 데이터를 생성하는 명령어인 make_regression()를 제공한다.
    
    - make_regression() 명령으로 만들어진 데이터는 종속 변수 y의 값이 독립변수 x벡터의 선형 조합인 선형 관계를 가진다.

#### 하나의 데이터

$ \large y=b+w^{T}x+\epsilon$

$w^{T}x$ : M차원 벡터

$y, b, \epsilon$ : 모두 스칼라 값

#### 여러개의 데이터

$\large y = b1 + Xw + \epsilon $

$X$ : $M \times N$차원 행렬

$w$ : M차원 벡터

$\epsilon$ : N차원 벡터

1 : N차원 1벡터
- - -

$\epsilon$
    - 잡음(noise), 오차(disturbance)
    - 종속변수 y에 영향을 미치는 요인 중에서 우리가 고려하지 않는 것들을 모두 합친것
    - 중심 극한의 정리에 따라 모든 영향의 합은 정규 분포를 따를 가능성이 높기 때문에 기댓값이 0인 정규 분포 확률 변수로 가정

$b$
    - bias, y절편
    - x가 0일때 항상 y가 0일수는 없으니까

### make_regression( )

- 7개 넣어서, 2개 나온다.

- 입력인수

- make_regression() 명령은 내부적으로 독립변수, 오차, 계수벡터를 확률적으로 생성
- 종속 변수 데이터 y벡터를 계산해서 X,y,w 값을 출력한다.

```python
from sklearn.datasets import make_regression
X, y, w = make_regression(n_samples=10, n_features=1, bias=0, noise=0, coef=True, random_state=0)
```

```python
X, y, w = make_regression(n_samples =50, n_features=1, bias=100, noise=10, coef=True, random_state=0)
plt.scatter(X,y,s=100)
plt.show()
```

독립 변수가 여러개인 경우

x가 feature이고, 명암이 
$x_{1}, x_{2}$ 값이 바뀌면 y의 분포가 바뀌는지를 봐야한다.

```python
import matplotlib as mpl

X, y, w = make_regression(n_samples=300, n_features=2, noise=10, 
                          coef=True, random_state=0)

plt.scatter(X[:,0], X[:,1], c=y, s=100, cmap=mpl.cm.bone)
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis("equal")
plt.show()
```
